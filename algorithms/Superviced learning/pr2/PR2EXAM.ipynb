```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['area_sqft', 'bedrooms', 'bathrooms', 'location_score',
       'property_age', 'distance_city_km', 'near_school', 'near_metro',
       'crime_rate_index']]
Y = df[['house_price_inr']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
df.shape
```

```python
X_test.shape
```

```python
X_train.shape
```

```python

from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()

# Fit and Transform Training Data
X_train_scaled = scaler.fit_transform(X_train)

# Transform Testing Data
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import GridSearchCV
```

```python
model = LinearRegression()
model.fit(X_train_scaled, Y_train)
```

```python
train_score = model.score(
    X_train_scaled,
    Y_train
)

print("Training Score :", train_score)
```

```python
test_score = model.score(
    X_test_scaled,
    Y_test
)

print("Testing Score :", test_score)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
model_mse = mean_squared_error(
    Y_test,
    y_pred
)
print("MSE :", model_mse)
```

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, Y_train)

ridge_train_score = ridge.score(X_train_scaled, Y_train)
ridge_test_score = ridge.score(X_test_scaled, Y_test)
ridge_y_pred = ridge.predict(X_test_scaled)
ridge_mse = mean_squared_error(Y_test, ridge_y_pred)
ridge_rmse = np.sqrt(ridge_mse)
ridge_mae = mean_absolute_error(Y_test, ridge_y_pred)
ridge_r2 = r2_score(Y_test, ridge_y_pred)

print("Ridge Regression (L2)")
print("Training Score :", ridge_train_score)
print("Testing Score :", ridge_test_score)
print("MSE :", ridge_mse)
print("RMSE :", ridge_rmse)
print("MAE :", ridge_mae)
print("R2 Score :", ridge_r2)
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_scaled, Y_train)

lasso_train_score = lasso.score(X_train_scaled, Y_train)
lasso_test_score = lasso.score(X_test_scaled, Y_test)
lasso_y_pred = lasso.predict(X_test_scaled)
lasso_mse = mean_squared_error(Y_test, lasso_y_pred)
lasso_rmse = np.sqrt(lasso_mse)
lasso_mae = mean_absolute_error(Y_test, lasso_y_pred)
lasso_r2 = r2_score(Y_test, lasso_y_pred)

print("Lasso Regression (L1)")
print("Training Score :", lasso_train_score)
print("Testing Score :", lasso_test_score)
print("MSE :", lasso_mse)
print("RMSE :", lasso_rmse)
print("MAE :", lasso_mae)
print("R2 Score :", lasso_r2)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
import pandas as pd

# Create Ridge Model
ridge = Ridge(alpha=1)

# Train Model
ridge.fit(X_train_scaled, Y_train)

# Prediction
y_pred_ridge = ridge.predict(X_test_scaled)

# Training Score
ridge_train = ridge.score(
    X_train_scaled,
    Y_train
)

# Testing Score
ridge_test = ridge.score(
    X_test_scaled,
    Y_test
)

# MSE
ridge_mse = mean_squared_error(
    Y_test,
    y_pred_ridge
)

# Print Results
print("===== Ridge Regression (L2) =====")

print("Training Score :", ridge_train)

print("Testing Score :", ridge_test)

print("MSE :", ridge_mse)

# Coefficients Table
ridge_coef = pd.DataFrame({

    'Feature': X.columns,
    'Coefficient': ridge.coef_

})

print("\nRidge Coefficients")

print(ridge_coef)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
```

```python
rf.fit(X_train, Y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
mae = mean_absolute_error(Y_test, y_pred)

mse = mean_squared_error(Y_test, y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(Y_test, y_pred)

print("MAE :", mae)

print("MSE :", mse)

print("RMSE :", rmse)

print("R2 Score :", r2)
```

```python
plt.figure(figsize=(25,10))

plot_tree(
    rf.estimators_[0],
    feature_names=X.columns,
    filled=True,
    rounded=True,
    max_depth=17
)

plt.show()
```

```python
for i, tree in enumerate(rf.estimators_):

    y_tree_pred = tree.predict(X_test)

    score = r2_score(Y_test, y_tree_pred)

    print(f"Tree {i+1} R2 Score : {score}")
```

```python
# Best Tree
best_score = 0

best_tree = None

for i, tree in enumerate(rf.estimators_):

    y_tree_pred = tree.predict(X_test)

    score = r2_score(Y_test, y_tree_pred)

    if score > best_score:

        best_score = score

        best_tree = i

print("Best Tree :", best_tree + 1)

print("Best R2 Score :", best_score)
```
