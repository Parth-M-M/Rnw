```python
import numpy as np
import pandas as pd
```

```python
filepath = "E:\Rnw\maths\Exam\student_dataset.csv"
df = pd.read_csv(filepath)
```

```python
df.describe()
```

```python
df['final_exam_pass'] = df['final_exam_pass'].map({"Pass": 1, "Fail": 0})
df['group_discussion'] = df['group_discussion'].map({"Yes": 1, "No": 0})
```

```python
P_pass = df['final_exam_pass'].mean()
P_study10 = (df['study_hours'] > 10).mean()
P_att80 = (df['attendance'] > 80).mean()

print("P(Pass):", P_pass)
print("P(Study >10 hrs):", P_study10)
print("P(Attendance >80%):", P_att80)
```
