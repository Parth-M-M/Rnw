```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv(r"E:\Rnw\maths\Exam\practical exam\employee_performance.csv")
print(df.head())
print(df.columns)
```

```python
mean_salary = df['Salary'].mean()
median_salary = df['Salary'].median()
mode_salary = df['Salary'].mode()[0]
print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)
```

```python
variance_projects = df['Projects_Completed'].var()
std_projects = df['Projects_Completed'].std()
print("Variance of Projects Completed:", variance_projects)
print("Standard Deviation of Projects Completed:", std_projects)
```

```python
promotion_prob = (df['Promotion_Status'] == 'Yes').mean()
print("Probability of Promotion:", promotion_prob)
```

```python
contingency_table = pd.crosstab(df['Department'], df['Promotion_Status'])
print(contingency_table)
```

```python
high_perf = df[df['Performance_Score (0-100)'] > 80]
conditional_prob = (high_perf['Promotion_Status'] == 'Yes').mean()

print("P(Promotion | Performance_Score > 80):", conditional_prob)
```

```python
plt.figure()
sns.histplot(df['Performance_Score (0-100)'], kde=True, stat="density")
plt.title("Performance Score Distribution with Gaussian Curve")
plt.xlabel("Performance Score")
plt.ylabel("Density")
plt.show()
```

```python
salary_skewness = stats.skew(df['Salary'])
salary_kurtosis = stats.kurtosis(df['Salary'])

print("Skewness of Salary:", salary_skewness)
print("Kurtosis of Salary:", salary_kurtosis)
```

```python
plt.figure()
stats.probplot(df['Projects_Completed'], dist="norm", plot=plt)
plt.title("Q-Q Plot of Projects Completed")
plt.show()
```

```python
vectors = df[['Projects_Completed', 'Working_Hours']].head(5).values

v1 = vectors[0]
v2 = vectors[1]
```

```python
dot_product = np.dot(v1, v2)
print("Dot Product:", dot_product)
```

```python
norm1 = np.linalg.norm(v1, ord=1)
norm2 = np.linalg.norm(v1)

print("Norm-1:", norm1)
print("Norm-2:", norm2)
```

```python
cos_theta = dot_product / (np.linalg.norm(v1) * np.linalg.norm(v2))
angle_rad = np.arccos(cos_theta)
angle_deg = np.degrees(angle_rad)

print("Angle between vectors (degrees):", angle_deg)
```
