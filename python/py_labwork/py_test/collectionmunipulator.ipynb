```python
students = []
```

```python



while True:
    print("""
    1. Add Student
    2. Display All Students
    3. Update Student Information
    4. Delete Student
    5. Display Subjects Offered
    6. Exit
    """)
    choice = input("Enter your choice: ")

    match choice:
        case "1":
            temp = []
            print("Enter student details:")
            sid = int(input("Student ID: "))
            temp.append(sid)
            temp.append(input("Name: "))
            temp.append(int(input("Age: ")))
            temp.append(input("Grade: "))
            temp.append(input("Date of Birth (YYYY-MM-DD): "))
            temp.append(input("Subjects (comma-separated): "))
            students.append(temp)
            print("Student added successfully!\n")

        case "2":
            if not students:
                print("No students to display.\n")
            for s in students:
                print(f"Student ID: {s[0]}")
                print(f"Name: {s[1]}")
                print(f"Age: {s[2]}")
                print(f"Grade: {s[3]}")
                print(f"Date of Birth: {s[4]}")
                print(f"Subjects: {s[5]}")
                print("-" * 30)

        case "3":
            sid = int(input("Enter Student ID to update: "))
            found = False
            for s in students:
                if s[0] == sid:
                    print("""
                    1. Name
                    2. Age
                    3. Grade
                    4. Date of Birth
                    5. Subjects
                    """)
                    field = input("Enter Option number to update: ")
                    match field:
                        case "1":
                            s[1] = input("Enter new name: ")
                        case "2":
                            s[2] = int(input("Enter new age: "))
                        case "3":
                            s[3] = input("Enter new grade: ")
                        case "4":
                            s[4] = input("Enter new date of birth (YYYY-MM-DD): ")
                        case "5":
                            s[5] = input("Enter new subjects (comma-separated): ")
                        case _:
                            print("Invalid input number.")
                    print("Student information updated.\n")
                    found = True
                    break
            if not found:
                print("Student ID not found.\n")

        case "4":
            sid = int(input("Enter Student ID to delete: "))
            for i, s in enumerate(students):
                if s[0] == sid:
                    students.pop(i)
                    print("Student deleted successfully.\n")
                    break
            else:
                print("Student ID not found.\n")

        case "5":
            subjects = set()
            for s in students:
                subjects.update(map(str.strip, s[5].split(",")))
            print("Subjects Offered:", ", ".join(sorted(subjects)) + "\n")

        case "6":
            print("Exiting program.")
            break

        case _:
            print("Invalid Input!\n")
```
