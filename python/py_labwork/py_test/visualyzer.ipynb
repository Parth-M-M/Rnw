```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Load the uploaded inventory file
inventory_path = "C:\Users\parth\Downloads\inventory_large.csv"
inventory_df = pd.read_csv(inventory_path)

# Generate 30 days of dates for one month (e.g., June 2025)
start_date = datetime(2025, 6, 1)
date_list = [start_date + timedelta(days=i) for i in range(30)]

# Simulate daily sales
sales_records = []
for date in date_list:
    # Randomly select 10 to 20 books to be sold each day
    num_sales_today = random.randint(10, 20)
    titles_today = inventory_df.sample(n=num_sales_today)

    for _, row in titles_today.iterrows():
        max_quantity_available = row['Quantity']
        # Randomly choose quantity sold (1 to min of 5 or available stock)
        quantity_sold = random.randint(1, min(5, max_quantity_available)) if max_quantity_available > 0 else 0
        total_revenue = quantity_sold * row['Price']
        sales_records.append({
            "Date": date.strftime("%Y-%m-%d"),
            "Title": row['Title'],
            "Quantity Sold": quantity_sold,
            "Total Revenue": round(total_revenue, 2)
        })

# Create sales DataFrame
sales_df = pd.DataFrame(sales_records)

# Save to CSV
sales_csv_path = "/mnt/data/sales_june_2025.csv"
sales_df.to_csv(sales_csv_path, index=False)

sales_df.head()
```

```python
import csv
import random
from faker import Faker
from datetime import datetime, timedelta

# Initialize Faker for generating fake data
fake = Faker()

# Generate inventory data
inventory_data = []
title_to_price = {}
for _ in range(500):
    title = fake.sentence(nb_words=3).title().replace('.', '')
    # Ensure unique titles
    while title in title_to_price:
        title = fake.sentence(nb_words=3).title().replace('.', '')
    author = fake.name()
    genre = random.choice(['Fiction', 'Non-Fiction', 'Science Fiction', 'Mystery', 'Romance', 'Thriller'])
    price = round(random.uniform(5.0, 50.0), 2)
    quantity = random.randint(1, 100)
    inventory_data.append([title, author, genre, price, quantity])
    title_to_price[title] = price

# Write inventory data to CSV
with open('inventory_data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Author', 'Genre', 'Price', 'Quantity'])
    writer.writerows(inventory_data)

# Generate date list from January 1, 2025, to July 3, 2025
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 7, 3)
date_list = [start_date + timedelta(days=x) for x in range((end_date - start_date).days + 1)]

# Generate sales data
sales_data = []
for date in date_list:
    num_sales = random.randint(0, 10)  # Number of books sold on this day
    if num_sales > 0:
        # Select unique titles for this day
        sold_titles = random.sample(list(title_to_price.keys()), num_sales)
        for title in sold_titles:
            quantity_sold = random.randint(1, 5)
            price = title_to_price[title]
            total_revenue = round(quantity_sold * price, 2)
            sales_data.append([date.strftime('%Y-%m-%d'), title, quantity_sold, total_revenue])

# Write sales data to CSV
with open('sales_data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Date', 'Title', 'Quantity Sold', 'Total Revenue'])
    writer.writerows(sales_data)
```

```python
def add_record(self, record_dict):
        new_record = pd.DataFrame([record_dict])
        self.data = pd.concat([self.data, new_record], ignore_index=True)
        print("Record added successfully.")
        self.save_data()
```

```python
def read_data(self, num_rows=5):
        print("\n--- First Few Rows ---")
        print(self.data.head(num_rows))
```

```python
def search_by_column(self, column_name, value):
        if column_name in self.data.columns:
            result = self.data[self.data[column_name] == value]
            print(f"\n--- Records where {column_name} == {value} ---")
            print(result)
        else:
            print("Column not found.")
```

```python
def update_record(self, search_col, search_val, update_col, new_val):
        if search_col in self.data.columns and update_col in self.data.columns:
            condition = self.data[search_col] == search_val
            self.data.loc[condition, update_col] = new_val
            print("Record updated successfully.")
            self.save_data()
        else:
            print("Column(s) not found.")
```

```python
def delete_record(self, column_name, value):
        if column_name in self.data.columns:
            self.data = self.data[self.data[column_name] != value]
            print(f"Records with {column_name} == {value} deleted.")
            self.save_data()
        else:
            print("Column not found.")
```
