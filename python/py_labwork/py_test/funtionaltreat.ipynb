```python
def factorial(n):
    if n == 1: 
        return 1
    return n * factorial(n - 1)  
def summary(*args):
    return len(my_list) , min(my_list) , max(my_list) , sum(my_list) , sum(my_list)/len(my_list)


print(" Welcome to the Data Analyzer and Transformer Program ")
my_list = []
while True :
    print(
        """
        Main menu:
        1.Input Data
        2.Display Data Summary (Built-in Functions)
        3.Calculate Factorial (Recursion)
        4.Filter Data by Threshold (Lambda Function)
        5.Sort Data
        6.Display Dataset Statistics (Return Multiple Values)
        7.Exit Program
        """)
    User_choice = int(input("Please Enter Your Choice : "))
    match User_choice :

        case 1 :
            user_data = (input("Enter data for a ID array (separated by spaces): ")) 
            my_list = [ x for x in user_data.split(" ")]
            for i in my_list :
                if i.isnumeric == True :
                    pass
                else :
                    print("Invalid Input please Re-Enter Numeric data (Enter 1 ) ")
                    break
            print("Data has been stored successfully!")
        case 2 :
            print("Data summary :")
            print(f" - Total element {len(my_list)}")
            print(f" - Minimun value: {min(my_list)}")
            print(f" - Maximum value: {max(my_list)}")
            print(f" - Sum of all Values : {sum(my_list)}")
            print(f" - Average  value : {sum(my_list)/len(my_list)}")
        case 3 :
            User_input = int(input("Enter a number to calculate its factorial:"))
            print(f"The factorial of {User_input} is {factorial(User_input)}")
        case 4 :
            user_threshold = int(input("Enter Threshold Number :"))
            my_below_list = list(filter(lambda x : x < user_threshold , my_list ))
            print(f"Numbers Below Threshold : {my_below_list}")
            my_above_list = list(filter(lambda x : x > user_threshold , my_list ))
            print(f"Numbers above Threshold : {my_above_list}")
        case 5 :
            print(
                """
                Choose sorting option:
                1. Ascending
                2. Descending
                """)
            user_sorting_choice = int(input("Enter your choice: "))
            match user_sorting_choice :
                case 1 :
                    my_assending = sorted(my_list)
                    print(my_assending)
                case 2 :
                    my_descending = sorted(my_list , reverse = True )
                    print(my_descending)

        case 6 :
            length , minimum , maximum , summing , avearage = summary(*my_list)
            print(f"""
            Dataset Statistics:
            - Total data points : {length}
            - Minimum Value : {minimum}
            - Maximum Value : {maximum}
            - sum of all Value : {summing}
            - Avearage : {avearage}
            """)
        case 7 :
            print(" Thank you for using the Data Analyzer and Transformer Program. Goodbye! ")
            break  
```
