```python
import math

def selector(no):
    if 0 < no <= 3:
        return 0.25
    elif 3 < no <= 6:
        return 0.5
    elif 6 < no <= 9:
        return 0.75
    elif 9 < no <= 12:
        return 1
    else:
        return 0
    
print("== Enter length of footing ==")
length_foot = int(input("Enter foot : "))
length_inch = int(input("Enter inch : "))
print("== Enter breath of footing ==")
breath_foot = int(input("Enter foot : "))
breath_inch = int(input("Enter inch : "))
print("== Enter spacing of footing ==")
mm_center = int(input("Enter center : "))
print("== Enter thickness of rod ==")
rod_mm = int(input("Enter rod mm : "))
print("== top & bottom mesh ==")
mesh = int(input("Enter no of mesh : "))
print("== total no of footing ==")
no_of_footing = int(input("Enter total no of footing : "))

non_length = selector(length_inch)
non_breath = selector(breath_inch)
total_length_foot = length_foot + non_length
total_breath_foot = breath_foot + non_breath
total_length_inch = (length_foot * 12) + length_inch
total_breath_inch = (breath_foot * 12) + breath_inch

length_spacing = math.ceil(total_length_inch / mm_center)
breath_spacing = math.ceil(total_breath_inch / mm_center)

no_of_rod_on_length = length_spacing * total_breath_foot
no_of_rod_on_breath = breath_spacing * total_length_foot

total = no_of_rod_on_breath + no_of_rod_on_length 

total_top_bottom = total * mesh

total_footing = total_top_bottom * no_of_footing

no_of_rod = math.ceil(total_footing / 39)

total_footing_weight = math.ceil((((rod_mm*rod_mm)/162)*12)*no_of_rod)
# Optional: Print results
print("Length Spacing (in rods):", length_spacing)
print("Breath Spacing (in rods):", breath_spacing)
print("Number of rods along length:", no_of_rod_on_length)
print("Number of rods along breath:", no_of_rod_on_breath)
print("total length of rod :",total)
print("total mesh :",total_top_bottom,"ft")
print("total footing :",total_footing,"ft")
print("no of rods :",no_of_rod)
print("total footing wieght :", total_footing_weight,"kg")
```

```python
from pydub import AudioSegment, effects

# === Input and Output Paths ===
input_path = "SATSANG 1.mp3"           # Replace with your file name
output_path = "SATSANG_1_filtered.mp3" # Output file

# === Load Audio ===
audio = AudioSegment.from_file(input_path, format="mp3")

# === Apply Filters ===
# Remove low-frequency hum (below ~150Hz)
filtered = audio.high_pass_filter(150)

# Remove high-frequency hiss (above ~8000Hz)
filtered = filtered.low_pass_filter(8000)

# Normalize volume for clarity
filtered = effects.normalize(filtered)

# === Export Processed Audio ===
filtered.export(output_path, format="mp3")

print(f"Filtered audio saved as {output_path}")
```
