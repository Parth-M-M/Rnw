```python
plt.figure()
sns.histplot(df['Age'], kde=True)
plt.title("Distribution of Age")
plt.show()

plt.figure()
sns.histplot(df['Income'], kde=True)
plt.title("Distribution of Income")
plt.show()

plt.figure()
sns.histplot(df['Purchases'], kde=True)
plt.title("Distribution of Purchases")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Gender', y='Purchases', data=df)
plt.title("Gender vs Purchases")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Churn', y='Income', data=df)
plt.title("Income vs Churn")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df[['Age', 'Income', 'Purchases']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(
    df[['Age', 'Income', 'Purchases', 'Churn']],
    hue='Churn'
)
plt.show()
```
